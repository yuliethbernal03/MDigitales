Los arrays n-diemnsionales se pueden modificar haciendo uso de la propiedad shape o del metodo reshape


````python
numpy.ones((2,3))
a=numpy.ones(10)
print(a)
print(a.shape)
a.shape=[2,5]
print(a)
# Con el metodo reshape
a= numpy.ones(10)
print(a.shape)
b= a.reshape((2,5))
print(b)
print(b.shape)
````

Para arrays n dimensionales se usa la siguiente estructura:

1. Se hace referencia a cada elemento del array con tantos indices, separados por comas, como dimensiones tiene el array.
2. La combinación de indices va entre corchetes tras el nombre del array.

```nombrearray[indexDim1,indexDim2,..., indexDimN]```

El operador : sustituye a todo el rango de indices posibles en la dimension en la que aparece


````python
a=numpy.linspace(0,1,5)
print(a)
print(a[:])
b=numpy.arange(4).reshape((2,2))
print(b)
print(b[:,1])
print(b[0,:])
````

````python
a=numpy.arange(24,dtype='int')
a=a.reshape((2,3,4))
print(a)
````

````python
print(a[0])
print(a[0,:,:])

print(a[0,0])
print(a[0,0,:])

print(a[0,0,0])

# Que pasa si escribo a[1,1:3,:2]
````

# Metodo concatenate
plot_22.jpg


````python
a=numpy.array([[1,2],[3,4]])
print(a)
print(a.shape)
print('--')
b=numpy.array([[5,6]]) # OJO
print(b)
print(b.shape)
print('----')
print(numpy.concatenate((a,b),axis=0)) # Pegado por filas
print(b.T.shape)
print(numpy.concatenate((a,b.T),axis=1)) # Pegado por columns
````

Por que los siguientes arrays don diferentes?

```a=numpy.array([1,2,3])```

```b=numpy.array([[1,2,3]])```

Son diferentes porque las dimensiones de uno y otro son diferentes. 

A la hora de concatenar el numero de dimensiones de los arrays deben ser las mismas



````python
a=numpy.array([1,2,3])
print(a.shape)
b=numpy.array([[1,2,3]])
print(b.shape)

c=numpy.zeros((3,3))
print(numpy.concatenate((b,c),axis=0))
````

````python
a=numpy.array([[5,6,1],[2,3,8]])
print(a)
print(numpy.amax(a))
print(numpy.amax(a,axis=0))
print(numpy.amax(a,axis=1))
print(numpy.percentile(a,25))
print(numpy.percentile(a,25,axis=0))
print(numpy.percentile(a,25,axis=1))
print(numpy.ptp(a))
print(numpy.ptp(a,axis=1))
````
